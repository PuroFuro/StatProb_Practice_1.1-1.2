```python
import matplotlib.pyplot as plt
import statistics
from collections import Counter

data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

def Mean(data):

    jumlah_nilai = sum(data) 
    jumlah_total_data = len(data)
    mean_value = jumlah_nilai / jumlah_total_data
    return mean_value

def Mode(data):

    mode_value = []
    counter = Counter(data)
    mode_data = dict(counter) 
    max_counter = max(list(counter.values())) 
    print(f"Max Counter: {max_counter}")

    for k, v in mode_data.items():
        if v == max(list(counter.values())):
            mode_value.append(k)
    return mode_value
    
def Median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    index_tengah = n//2 

    if n % 2 == 0:
        print(f"Jumlah total data Genap")
        median1 = sorted_data[index_tengah]
        median2 = sorted_data[index_tengah - 1]
        print(f"Nilai tengah: {median1} {median2}")
        median_value = (median1 + median2) / 2
        
    else:
        print(f"Jumlah total data Ganjil")
        median_value = sorted_data[index_tengah]
        print(f"Nilai tengah: {median_value}")
    
    return median_value

quest = input("apa yang ingin dicari?\n(a) Mean\n(b) Median\n(c) Modus\n= ")
dialog = ''
datatype = []
if quest == 'a':

    datatype = Mean(data)
    dialog = 'Mean'
    print("(Jika pakai library) mean = " + str(statistics.mean(data)))

elif quest == 'b':

    datatype = Median(data)
    dialog = 'Median'
    print("(Jika pakai library) median = " + str(statistics.median(data)))

elif quest == 'c':
    
    datatype = Mode(data)
    dialog = 'Modus'
    print("(Jika pakai library) mode = " + str(statistics.mode(data)))

print(f'Jadi ' + dialog + ' dari data tersebut adalah ' + str(datatype))

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

if quest != 'c':
    plt.axvline(x=datatype, color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {datatype:.2f}')
else:
    angka = 1
    for numbers in datatype:
        plt.axvline(x=numbers, color='red', linestyle='dashed', linewidth=1, label=f'Modus ' + str(angka) + f' = {numbers:.2f}')
        angka += 1


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(str(dialog))

plt.legend()
plt.show()
```
